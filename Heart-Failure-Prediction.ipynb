```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("heart_failure_clinical_records_dataset.csv")
dataset.head(3)
```

```python
dataset = dataset.drop(columns='time' , axis=1)
```

```python
dataset.isnull().sum()
```

```python
dataset.duplicated().sum()
```

```python
dataset.describe()
```

```python
x = dataset.drop('DEATH_EVENT' , axis=1)
y = dataset['DEATH_EVENT']
```

```python
y.shape , x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape , y_train.shape , X_test.shape , y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm= MinMaxScaler()
```

```python
x_train = mm.fit_transform(X_train)
x_test  = mm.transform(X_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
lr.score(x_train , y_train) , lr.score(x_test , y_test)
```

```python
# Remove the extra feature (assuming the extra feature is the last one)
input_data = np.array([[75.0, 0, 582, 0, 20, 1, 265000.00, 1.9, 130, 1, 0]])

# Transform the data using MinMaxScaler
scaled_input_data = mm.transform(input_data)

# Make a prediction
prediction = lr.predict(scaled_input_data)

print(prediction)
```

```python
import joblib

# Save the model and scaler
joblib.dump(lr, 'model.pkl')
joblib.dump(mm, 'scaler.pkl')

print("Model and scaler saved successfully!")
```
